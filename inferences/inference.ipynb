```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
plt.rcParams["figure.dpi"] = 300
```

```python

# initalize array to hold dataframe
dataframes = []

# reading through the 4 heuristics
for i in range(1, 5):
    filename = f"../results/heuristic_data_{i}.csv"
    if os.path.exists(filename):  # checking if file exists
        df = pd.read_csv(filename).drop(columns=["Unnamed: 0"])
        dataframes.append(df)
    else:
        #error handling if file not found
        print(f"File '{filename}' not found.")


```

```python
# Plot the arrays
plt.plot(dataframes[0].index, dataframes[0]["fringe_size"], label='fringe size heuristic 1', alpha=0.5)
plt.plot(dataframes[1].index, dataframes[1]["fringe_size"], label='fringe size heuristic 2', alpha=0.5)
plt.plot(dataframes[2].index, dataframes[2]["fringe_size"], label='fringe size heuristic 3', alpha=0.5)
plt.plot(dataframes[3].index, dataframes[3]["fringe_size"], label='fringe size heuristic 4', alpha=0.5)

# Add labels and legend
plt.xlabel('Scenarios')
plt.ylabel('Fringe Size')
plt.legend()


plt.ylim(-15000, 35000)
# Show the plot
plt.show()
```

```python
# Plot the arrays
plt.plot(dataframes[0].index, dataframes[0]["expanded_node"], label='expanded node heuristic 1', alpha=0.5)
plt.plot(dataframes[1].index, dataframes[1]["expanded_node"], label='expanded node heuristic 2', alpha=0.5)
plt.plot(dataframes[2].index, dataframes[2]["expanded_node"], label='expanded node heuristic 3', alpha=0.5)
plt.plot(dataframes[3].index, dataframes[3]["expanded_node"], label='expanded node heuristic 4', alpha=0.5)

# Add labels and legend
plt.xlabel('Scenarios')
plt.ylabel('Expanded Node')
plt.legend()


plt.ylim(-50000, 300000)
# Show the plot
plt.show()
```
